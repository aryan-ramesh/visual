```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step1")
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.set_title('Multiple Lines')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitdue')
plt.show();
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',color='yellow')
ax.plot(x,np.cos(x),label='cos',color='blue')
ax.plot(x,np.tan(x),label='tan',color='lime')
ax.set_title('Multiple Lines')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitdue')
ax.legend()
plt.show();
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='tan',linestyle='dashdot')
ax.set_title('Multiple Lines')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitdue')
ax.legend()
plt.show();
```
